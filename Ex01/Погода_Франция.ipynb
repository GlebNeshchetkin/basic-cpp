```
def MakeList (List:list, file_address:str):
  with open(file_address, 'r') as data_file:
    csvreader = csv.reader(data_file, delimiter=';')
    for row in csvreader:
        List.append(float(row[0]))
```

```
def SummarizeTemp (List:list):
  sum=0
  for i in range(12):
   sum+=List[i]
  print("The average temperature in {} over the year is about {:.2f} degree".format(List[12],sum/12))
```

```
def SummarizeRain (List:list, sum:int)->int:
  for element in List[:12]:
   sum+=element   
  print("Annual rainfall in {} is {:.0f} mm".format(List[12],sum))
  return sum
```

```
import csv
Paris=[]
MakeList(Paris,'/content/Paris_t_.csv')
Paris.append('Paris')
Lille=[]
MakeList(Lille,'/content/Lille_t_.csv')
Lille.append('Lille')
Nice=[]
MakeList(Nice,'/content/Nice_t_.csv')
Nice.append('Nice')
SummarizeTemp(Paris)
SummarizeTemp(Lille)
SummarizeTemp(Nice)
ParisO=[]
MakeList(ParisO,'/content/Paris_o_.csv')
ParisO.append('Paris')
LilleO=[]
MakeList(LilleO,'/content/Lille_o_.csv')
LilleO.append('Lille')
NiceO=[]
MakeList(NiceO,'/content/Nice_o_.csv')
NiceO.append('Nice')
sum1=0 
sum2=0 
sum3=0
sum1=SummarizeRain(ParisO,sum1)
sum2=SummarizeRain(LilleO, sum2)
sum3=SummarizeRain(NiceO, sum3)
```

```
if sum1>sum2 and sum1>sum3:
 print("Most rainfall is observed in Paris")
if sum2>sum1 and sum2>sum3:
 print("Most rainfall is observed in Lille")
if sum3>sum2 and sum3>sum1:
 print("Most rainfall is observed in Nice")
```
